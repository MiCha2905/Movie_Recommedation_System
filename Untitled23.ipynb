```python
GENRES = {
    "Action": 28,
    "Comedy": 35,
    "Drama": 18,
    "Thriller": 53
}
```

```python
def fetch_movies_by_genre(genre_name, genre_id, pages=5):
    movies = []

    for page in range(1, pages + 1):
        url = "https://api.themoviedb.org/3/discover/movie"
        params = {
            "api_key": API_KEY,
            "with_genres": genre_id,
            "language": "en-US",
            "sort_by": "popularity.desc",
            "page": page
        }

        response = requests.get(url, params=params)
        data = response.json()

        for item in data["results"]:
            movies.append({
                "title": item["title"],
                "year": item["release_date"][:4] if item["release_date"] else None,
                "rating": item["vote_average"],
                "genre": genre_name,
                "overview": item["overview"]
            })

    return movies
```

```python
all_movies = []

for genre_name, genre_id in GENRES.items():
    print(f"Fetching {genre_name} movies...")
    genre_movies = fetch_movies_by_genre(genre_name, genre_id, pages=5)
    print(f"Collected {len(genre_movies)} {genre_name} movies")
    all_movies.extend(genre_movies)
```

```python
df = pd.DataFrame(all_movies)

# Remove duplicates
df.drop_duplicates(subset=["title", "year"], inplace=True)

df.to_csv("movies.csv", index=False)

print("TOTAL MOVIES IN DATASET:", len(df))
df.head()
```

```python
indices = pd.Series(df.index, index=df["title"].str.lower())
indices.head()
```

```python
def recommend_movies(movie_title, top_n=5):
    movie_title = movie_title.lower()

    if movie_title not in indices:
        return ["Movie not found in database."]

    idx = indices[movie_title]
    similarity_scores = list(enumerate(cosine_sim[idx]))

    similarity_scores = sorted(similarity_scores, key=lambda x: x[1], reverse=True)
    similarity_scores = similarity_scores[1:top_n+1]

    recommended_indices = [i[0] for i in similarity_scores]

    return df["title"].iloc[recommended_indices].tolist()
```

```python
import re

def clean_user_input(text):
    text = text.lower()
    text = re.sub(r"[^a-zA-Z\s]", "", text)
    return text.strip()
```

```python
def detect_intent(user_input):
    if "like" in user_input or "similar" in user_input:
        return "similar_movie"
    elif "action" in user_input or "comedy" in user_input or "drama" in user_input or "thriller" in user_input:
        return "genre"
    elif "rating" in user_input or "above" in user_input:
        return "rating"
    else:
        return "general"
```

```python
def extract_movie_name(user_input):
    match = re.search(r"like (.+)", user_input)
    if match:
        return match.group(1).title()
    return None
```

```python
def recommend_by_genre(genre, top_n=5):
    genre = genre.lower()
    filtered = df[df["genre"].str.lower().str.contains(genre)]
    return filtered.sort_values(by="rating", ascending=False)["title"].head(top_n).tolist()
```

```python
def recommend_by_rating(min_rating, top_n=5):
    filtered = df[df["rating"] >= min_rating]
    return filtered.sort_values(by="rating", ascending=False)["title"].head(top_n).tolist()
```

```python
def recommend_by_rating(min_rating, top_n=5):
    df["rating"] = pd.to_numeric(df["rating"], errors="coerce")

    filtered = df[df["rating"] >= min_rating]

    if filtered.empty:
        return [f"No movies found with rating above {min_rating}"]

    filtered = filtered.sort_values(by="rating", ascending=False)

    return filtered[["title", "rating"]].head(top_n).values.tolist()
```

```python
def chatbot_response(user_input):
    raw_input = user_input.lower()   # keep raw input for number extraction
    clean_input = clean_user_input(user_input)

    intent = detect_intent(clean_input)

    if intent == "similar_movie":
        movie = extract_movie_name(raw_input)
        if movie:
            return recommend_movies(movie)
        else:
            return ["Please mention a movie name."]

    elif intent == "genre":
        for g in ["action", "comedy", "drama", "thriller"]:
            if g in clean_input:
                return recommend_by_genre(g)
        return ["Please specify a genre."]

    elif intent == "rating":
        match = re.search(r"\d+(\.\d+)?", raw_input)
        if match:
            min_rating = float(match.group())
            return recommend_by_rating(min_rating)
        else:
            return ["Please mention a rating value."]

    else:
        return df.sort_values(by="rating", ascending=False)["title"].head(5).tolist()
```

```python
import ipywidgets as widgets
from IPython.display import display, clear_output
import pandas as pd
import pickle
import re
from sklearn.metrics.pairwise import cosine_similarity
```

```python
df = pd.read_csv("movies_cleaned.csv")

tfidf = pickle.load(open("tfidf_vectorizer.pkl", "rb"))
tfidf_matrix = pickle.load(open("tfidf_matrix.pkl", "rb"))

cosine_sim = cosine_similarity(tfidf_matrix, tfidf_matrix)
indices = pd.Series(df.index, index=df["title"].str.lower())
```

```python
def normalize(text):
    return re.sub(r"[^a-zA-Z0-9\s]", "", text.lower()).strip()

# Create normalized title index
normalized_indices = {
    normalize(title): idx
    for idx, title in enumerate(df["title"])
}
```

```python
def clean_user_input(text):
    return re.sub(r"[^a-zA-Z\s]", "", text.lower()).strip()

def recommend_movies(movie_query, top_n=5, min_rating=7):
    query_norm = normalize(movie_query)

    for title_norm, idx in normalized_indices.items():
        if query_norm in title_norm:
            base_genre = df.loc[idx, "genre"].lower()

            # 1️⃣ Similarity-based candidates
            scores = list(enumerate(cosine_sim[idx]))
            scores = sorted(scores, key=lambda x: x[1], reverse=True)

            recommendations = []
            for i, score in scores[1:]:
                if (base_genre in df.loc[i, "genre"].lower() and
                    df.loc[i, "rating"] >= min_rating):
                    recommendations.append(df.loc[i, "title"])
                if len(recommendations) == top_n:
                    break

            # 2️⃣ Fallback if similarity is weak
            if len(recommendations) < 3:
                fallback = (
                    df[df["genre"].str.lower().str.contains(base_genre)]
                    .sort_values(by="rating", ascending=False)
                    ["title"]
                    .head(top_n)
                    .tolist()
                )
                return fallback

            return recommendations

    return ["Movie not found in database."]





def recommend_by_genre(genre):
    filtered = df[df["genre"].str.lower().str.contains(genre)]
    return filtered.sort_values(by="rating", ascending=False)["title"].head(5).tolist()

def recommend_by_rating(min_rating):
    df["rating"] = pd.to_numeric(df["rating"], errors="coerce")
    filtered = df[df["rating"] >= min_rating]
    if filtered.empty:
        return ["No movies found."]
    return filtered.sort_values(by="rating", ascending=False)[["title", "rating"]].head(5).values.tolist()

def chatbot_response(user_input):
    raw = user_input.lower()
    clean = clean_user_input(user_input)

    if "like" in clean:
        movie = raw.split("like")[-1].strip()
        return recommend_movies(movie)

    for g in ["action", "comedy", "drama", "thriller"]:
        if g in clean:
            return recommend_by_genre(g)

    if "rating" in clean or "above" in clean:
        match = re.search(r"\d+(\.\d+)?", raw)
        if match:
            return recommend_by_rating(float(match.group()))
        return ["Please mention a rating value."]

    return df.sort_values(by="rating", ascending=False)["title"].head(5).tolist()
```

```python
def get_movie_genre(movie_query):
    query_norm = normalize(movie_query)
    for title_norm, idx in normalized_indices.items():
        if query_norm in title_norm:
            return df.loc[idx, "genre"].lower()
    return None
```

```python
chat_history = []

input_box = widgets.Text(
    placeholder='Type your message here...',
    description='You:',
    layout=widgets.Layout(width='80%')
)

send_button = widgets.Button(description="Send")

output_area = widgets.Output()

def handle_send(b):
    user_text = input_box.value
    input_box.value = ""

    if user_text.strip() == "":
        return

    bot_response = chatbot_response(user_text)

    with output_area:
        print("You:", user_text)
        print("Bot:")
        for r in bot_response:
            print(" -", r)
        print("-" * 40)

send_button.on_click(handle_send)

display(input_box, send_button, output_area)
```
